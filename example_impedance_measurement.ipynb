```python
import time
import numpy as np
import matplotlib.pyplot as plt
import visa 
from agilent4284a import Agilent4284A
rm = visa.ResourceManager()
rm.list_resources()
```

```python
lcr = Agilent4284A('GPIB0::3::INSTR')
lcr.set_impedance_range(10)
```

```python
freq = np.logspace(1.3011, 4, 100)
impedance = []
for f in freq:
    print(f)
    lcr.set_frequency(f)
    
    impedance.append(lcr.get_impedance())
```

```python
plt.plot(impedance)
plt.show()
```
